```python
import dhlab.nbtext as nb
```

## Bygg korpus

```python
korpus = nb.book_corpus(words = ["tyske", "tysk"], ddk = "839%", lang="nob", limit = 20000, period = (1945, 1980))
```

```python
korpus
```

## Refine


```python
korpus2 = nb.frame(
    nb.refine_book_urn(words =['soldat', 'soldater'], urns = list(korpus["urn"]), limit=3000),
"urn author title year".split())
```

```python
korpus2
```

## Test en konk

```python
nb.urn_concordance(word =['soldater', 'soldat'], urns = list(korpus2["urn"])[11:50])
```
